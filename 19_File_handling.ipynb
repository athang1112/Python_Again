6. A menu based program to create file and to count number of lines in this file

```python
import os.path 
from os import path

while True:
    print('------------Main Menu-----------')
    print('1. Display text file')
    print('2. Add line to a text file')
    print('3. Count the number of lines')
    print('4. Count words in file')
    print('5. Count spaces and characters')
          
    choice=input("Enter your choice (other to exit): ")

    print()
    fh=open('file1.txt','a+')           #file handle

    if int(choice)==1:
        fh.seek(0)
        contents=fh.readlines()
        for i in range(0,len(contents)):
            print(contents[i],end='')

    elif int(choice)==2:
        option='yes'
        while option in 'YESYesyes':
            a=input("Enter the line to insert in file: ")
            fh.write(a)
            fh.write('\n')
            option=input('Add another line (yes/no)?: ')

    elif int(choice)==3:
        if path.exists('file1.txt'):
            fh.seek(0)
            count=0
            for line in fh:
                count+=1
            print("The no of lines are: ",count)
        else:
            print('File does not exist')

    elif int(choice)==4:            #count words
        if path.exists('file1.txt'):
            fh.seek(0)
            count=0
            for line in fh:
                w=line.split()
                count+=len(w)
            print('Total words in file are: ',count)
        else:
            print("File does not exist")
    elif int(choice)==5:           #count spaces and characters
        if path.exists('file1.txt'):
            fh.seek(0)
            spaces=chars=0
            x=fh.read()
            for c in range(len(x)):
                if x[c]==' ':
                    spaces+=1
                else:
                    chars+=1
            print("Total %d spaces and %d characters in file"%(spaces,chars))
        else:
            print('File does not exist')
    else:
        print("Exiting...")
        break
```

```python
import os
from os import path

filename = "file1.txt"

while True:
    print("\n------------ Main Menu -----------")
    print("1. Display text file")
    print("2. Add line to a text file")
    print("3. Count the number of lines")
    print("4. Count words in file")
    print("5. Count spaces and characters")

    choice = input("Enter your choice (other to exit): ")

    if not choice.isdigit():
        print("Exiting...")
        break

    choice = int(choice)

    # 1️⃣ Display file
    if choice == 1:
        if path.exists(filename):
            with open(filename, "r") as fh:
                print("\nFile Content:\n")
                print(fh.read())
        else:
            print("File does not exist.")

    # 2️⃣ Add line
    elif choice == 2:
        with open(filename, "a") as fh:
            while True:
                line = input("Enter line to insert: ")
                fh.write(line + "\n")
                option = input("Add another line (yes/no)?: ")
                if option.lower() != "yes":
                    break

    # 3️⃣ Count lines
    elif choice == 3:
        if path.exists(filename):
            with open(filename, "r") as fh:
                count = len(fh.readlines())
                print("Total number of lines:", count)
        else:
            print("File does not exist.")

    # 4️⃣ Count words
    elif choice == 4:
        if path.exists(filename):
            with open(filename, "r") as fh:
                words = fh.read().split()
                print("Total words in file:", len(words))
        else:
            print("File does not exist.")

    # 5️⃣ Count spaces & characters
    elif choice == 5:
        if path.exists(filename):
            with open(filename, "r") as fh:
                text = fh.read()
                spaces = text.count(" ")
                chars = len(text.replace(" ", ""))
                print("Total spaces:", spaces)
                print("Total characters (excluding spaces):", chars)
        else:
            print("File does not exist.")

    else:
        print("Exiting...")
        break
```
