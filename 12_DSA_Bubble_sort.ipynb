## Bubble Sort

```python
#while loop
def bubble(l):
    i=0
    while i<len(l):
        j=0
        while j<len(l)-i-1:
            if l[j]>l[j+1]:
                l[j],l[j+1]=l[j+1],l[j]
            j+=1
        i+=1
    print(l)

bubble([4,1,7,2,8])


#manual swap 
# def bubble(l):
#     i = 0
#     while i < len(l):
#         j = 0
#         while j < len(l) - i - 1:
#             if l[j] > l[j + 1]:
#                 temp = l[j]
#                 l[j] = l[j + 1]
#                 l[j + 1] = temp
#             j += 1
#         i += 1
#     print(l)

# l = eval(input("Enter list: "))
# bubble(l)
```

What does len(l) - i - 1 mean?

It tells the last index up to which we should compare in each pass.

ðŸ‘‰ In bubble sort, after every pass, the largest element moves to the end
So we donâ€™t need to compare it again

```python
#descending order
def bubble(l):
    for i in range(len(l)):
        for j in range(len(l)-i-1):
            if l[j]<l[j+1]:
                l[j],l[j+1]=l[j+1],l[j]
    print(l)

bubble([6,2,9,4,1])
```

```python
#counting number of swaps
def bubble(l):
    swaps=0
    for i in range(len(l)):
        for j in range(len(l)-i-1):
            if l[j]>l[j+1]:
                l[j],l[j+1]=l[j+1],l[j]
                swaps+=1
    print("Sorted List:",l)
    print("Swaps:",swaps)
bubble([6,1,8,3,2,4])
```

```python
#count comparisons
def bubble(l):
    comp=0
    for i in range(len(l)):
        for j in range(len(l)-i-1):
            comp+=1
            if l[j]>l[j+1]:
                l[j],l[j+1]=l[j+1],l[j]
    print("Sorted List:",l)
    print("Comparisions:",comp)
bubble([6,8,4,5,2,0])
```

```python
#sort only first n elements
def bubble(l,n):
    for i in range(n):
        for j in range(n-i-1):
            if l[j]>l[j+1]:
                l[j],l[j+1]=l[j+1],l[j]
    print(l)

bubble([6,3,8,2,5,1,0],4)
```

```python
#reverse loop style
def bubble(l):
    for i in range(len(l) - 1, 0, -1):
        for j in range(i):
            if l[j] > l[j + 1]:
                l[j], l[j + 1] = l[j + 1], l[j]
    print(l)

bubble([4, 6, 1, 9, 2])
```

```python
#early stop + pass count
def bubble(l):
    for i in range(len(l)):
        swap=False
        for j in range(len(l)-i-1):
            if l[j]>l[j+1]:
                l[j],l[j+1]=l[j+1],l[j]
                swap=True
        print("After pass: ",i+1,":",l)
        if not swap:
            break
    print("Final:",l)

bubble([5,8,6,3,4,2])
```

```python
#track last swap index
def bubble(l):
    n = len(l)
    while n > 0:
        last = 0
        for j in range(n - 1):
            if l[j] > l[j + 1]:
                l[j], l[j + 1] = l[j + 1], l[j]
                last = j + 1
        n = last
    print(l)

bubble([6, 3, 5, 2, 1, 4])
```

```python
#recursive version
def bubble(l, n):
    if n == 1:
        return
    for j in range(n - 1):
        if l[j] > l[j + 1]:
            l[j], l[j + 1] = l[j + 1], l[j]
    bubble(l, n - 1)

l = [7, 4, 6, 2, 5]
bubble(l, len(l))
print(l)
```

```python
# sort only even indexes
def bubble(l):
    for i in range(len(l)):
        for j in range(0,len(l) - i - 1, 2):
            if l[j] > l[j + 1]:
                l[j], l[j + 1] = l[j + 1], l[j]
    print(l)

bubble([9, 4, 7, 2, 6, 1])
```

```python
#count passes, swaps, comparisons
def bubble(l):
    passes = swaps = comps = 0
    for i in range(len(l)):
        passes += 1
        for j in range(len(l) - i - 1):
            comps += 1
            if l[j] > l[j + 1]:
                l[j], l[j + 1] = l[j + 1], l[j]
                swaps += 1
    print(l)
    print("Passes:", passes, "Swaps:", swaps, "Comparisons:", comps)

bubble([5, 4, 3, 2, 1])
```

```python
#while loop + reverse indexing
def bubble(l):
    i = len(l) - 1
    while i > 0:
        j = 0
        while j < i:
            if l[j] > l[j + 1]:
                l[j], l[j + 1] = l[j + 1], l[j]
            j += 1
        i -= 1
    print(l)

bubble([8, 1, 6, 3, 7])
```

```python
#strings (lexicographical)
def bubble(l):
    for i in range(len(l)):
        for j in range(len(l) - i - 1):
            if l[j] > l[j + 1]:
                l[j], l[j + 1] = l[j + 1], l[j]
    print(l)

bubble(["pear", "apple", "orange", "banana"])
```

```python
#tuple data (by second value)
def bubble(l):
    for i in range(len(l)):
        for j in range(len(l) - i - 1):
            if l[j][1] > l[j + 1][1]:
                l[j], l[j + 1] = l[j + 1], l[j]
    print(l)

bubble([(1, 5), (2, 3), (3, 8), (4, 1)])
```

```python
#partial range (start to end)
def bubble(l, start, end):
    # Bubble sort only from index start to end (inclusive)
    for i in range(start, end):
        for j in range(start, end - (i - start)):
            if l[j] > l[j + 1]:
                l[j], l[j + 1] = l[j + 1], l[j]
    print(l)

bubble([9, 7, 5, 3, 1, 2, 4, 6], 2, 6)
```
