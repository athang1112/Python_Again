## Insertion Sort

```python
def insertion(l):
    for i in range(1, len(l)):
        key = l[i] #8 #3
        j = i - 1 #5 #8
        while j >= 0 and key < l[j]:  #3<8 
            l[j + 1] = l[j] #8
            j -= 1 #
        l[j + 1] = key#3
    print("List after sorting is", l)

insertion([5,8,3,6,2,1])
```

```python
#no while loop inside
def insertion_for(a):
    for i in range(1, len(a)):
        key = a[i]
        pos = i  # track correct position

        for j in range(i - 1, -1, -1):
            if a[j] > key:
                a[j + 1] = a[j]
                pos = j
            else:
                break

        a[pos] = key

    print(a)

insertion_for([5, 8, 3, 6, 2])
```

```python
#descending order 
def insertion_desc(a):
    for i in range(1, len(a)):
        key = a[i]
        j = i - 1
        while j >= 0 and a[j] < key:
            a[j + 1] = a[j]
            j -= 1
        a[j + 1] = key
    print(a)

insertion_desc([5, 8, 3, 6, 2])
```

```python
#Insertion sort with step-by-step output (dry run)
def insertion_trace(a):
    for i in range(1, len(a)):
        key = a[i]
        j = i - 1
        while j >= 0 and a[j] > key:
            a[j + 1] = a[j]
            j -= 1
        a[j + 1] = key
        print("After pass", i, ":", a)

insertion_trace([5, 8, 3, 6, 2])
```

```python
# Insertion sort with INNER while loop trace
def insertion_trace_inner(a):
    for i in range(1, len(a)):
        key = a[i]
        j = i - 1
        print("\nPass", i, ": key =", key)

        while j >= 0 and a[j] > key:
            print(f"  while: a[{j}] = {a[j]} > key ({key}) → shift")
            a[j + 1] = a[j]
            print("   array:", a)
            j -= 1

        a[j + 1] = key
        print(f"  insert key at position {j+1}")
        print("  array after pass:", a)

insertion_trace_inner([5, 8, 3, 6, 2])
```

```python
#Insertion sort using recursion
def insertion_recursive(a, n):
    if n <= 1:
        return
    insertion_recursive(a, n - 1)
    key = a[n - 1]
    j = n - 2
    while j >= 0 and a[j] > key:
        a[j + 1] = a[j]
        j -= 1
    a[j + 1] = key

arr = [5, 8, 3, 6, 2]
insertion_recursive(arr, len(arr))
print(arr)
```

```python
#sort without extra variables (index practice)
def insertion_no_temp(a):
    for i in range(1, len(a)):
        j = i
        while j > 0 and a[j] < a[j - 1]:
            a[j], a[j - 1] = a[j - 1], a[j]
            j -= 1
    print(a)

insertion_no_temp([5, 8, 3, 6, 2])
```

```python
#sort with comparison + shift counters
def insertion_with_counts(a):
    comps = shifts = 0

    for i in range(1, len(a)):
        key = a[i]
        j = i - 1

        while j >= 0:
            comps += 1
            if a[j] > key:
                a[j + 1] = a[j]
                shifts += 1
                j -= 1
            else:
                break

        a[j + 1] = key

    print("Sorted:", a)
    print("Comparisons:", comps)
    print("Shifts:", shifts)

insertion_with_counts([5, 8, 3, 6, 2])
```

```python
#sort on subarray (used in hybrid sorts)
def insertion_subarray(a, start, end):
    for i in range(start + 1, end + 1):
        key = a[i]
        j = i - 1
        while j >= start and a[j] > key:
            a[j + 1] = a[j]
            j -= 1
        a[j + 1] = key

arr = [9, 7, 5, 3, 8, 6, 2]
insertion_subarray(arr, 1, 5)
print(arr)
```

```python
#Binary Insertion Sort (advanced optimization)

#Reduces comparisons using binary search.
def binary_insertion_sort(a):
    for i in range(1, len(a)):
        key = a[i]
        low, high = 0, i - 1

        while low <= high:
            mid = (low + high) // 2
            if a[mid] > key:
                high = mid - 1
            else:
                low = mid + 1

        for j in range(i - 1, low - 1, -1):
            a[j + 1] = a[j]

        a[low] = key

arr = [5, 8, 3, 6, 2]
binary_insertion_sort(arr)
print(arr)
```

```python
#Insertion Sort with Sentinel (no boundary check)
def insertion_with_sentinel(a):
    a.insert(0, float('-inf'))

    for i in range(2, len(a)):
        key = a[i]
        j = i - 1
        while a[j] > key:
            a[j + 1] = a[j]
            j -= 1
        a[j + 1] = key

    a.pop(0)

arr = [5, 8, 3, 6, 2]
insertion_with_sentinel(arr)
print(arr)
```

```python
#Stable Insertion Sort (explicit stability handling)
def stable_insertion_sort(a):
    for i in range(1, len(a)):
        key = a[i]
        j = i - 1

        while j >= 0 and a[j] > key:
            j -= 1

        a.pop(i)
        a.insert(j + 1, key)

arr = [3, 5, 3, 2, 5]
stable_insertion_sort(arr)
print(arr)
```

```python
# Sort with Swap-based Inner Loop

# (Contrast with shift-based version)

def insertion_swap_based(a):
    for i in range(1, len(a)):
        j = i
        while j > 0 and a[j] < a[j - 1]:
            a[j], a[j - 1] = a[j - 1], a[j]
            j -= 1

arr = [5, 8, 3, 6, 2]
insertion_swap_based(arr)
print(arr)
```

```python
# Insertion Sort with Early Break Counter

# (Shows adaptive nature clearly
def insertion_early_break(a):
    breaks = 0

    for i in range(1, len(a)):
        key = a[i]
        j = i - 1
        while j >= 0 and a[j] > key:
            a[j + 1] = a[j]
            j -= 1
        if j == i - 1:
            breaks += 1
        a[j + 1] = key

    print("Sorted:", a)
    print("Early breaks:", breaks)

insertion_early_break([1, 2, 3, 6, 5])
```

```python
# Sort Using bisect (library-assisted)

# (Conceptually binary insertion)

import bisect

def insertion_bisect(a):
    for i in range(1, len(a)):
        key = a.pop(i)
        pos = bisect.bisect_left(a, key, 0, i)
        a.insert(pos, key)

arr = [5, 8, 3, 6, 2]
insertion_bisect(arr)
print(arr)
```

```python
# Sort on Nearly Sorted Data

# (Real-world performance tuning)

def insertion_nearly_sorted(a):
    for i in range(1, len(a)):

        # safe optimization
        if a[i] >= a[i - 1]:
            continue

        key = a[i]
        j = i - 1
        while j >= 0 and a[j] > key:
            a[j + 1] = a[j]
            j -= 1
        a[j + 1] = key

arr = [1, 2, 4, 3, 5, 6]
insertion_nearly_sorted(arr)
print(arr)

```
