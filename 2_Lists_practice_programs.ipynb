```python
#Matrix Multiplication (User Input)(only if dimensions are compatible)
print("Enter dimensions for matrix A ")
rowsA=int(input("Enter rows for matrix A: "))
colsA=int(input("Enter columns for matrix A: "))

print("\nEnter dimensions for matrix B ")
rowsB=int(input("Enter rows for matrix B: "))
colsB=int(input("Enter columns for matrix B: "))

if colsA!=rowsB:
    print("Matrix multiplication is not possible")
else:
    
    print("\nElements for Matrix A: ")
    A=[]
    for i in range(rowsA):
        row=[]
        for j in range(colsA):
            ele=int(input(f'Enter element for ({i+1},{j+1})'))
            row.append(ele)
        A.append(row)

    print("\nElements for Matrix B: ")
    B=[]
    for i in range(rowsB):
        row=[]
        for j in range(colsB):
            ele=int(input(f'Enter element for ({i+1},{j+1})'))
            row.append(ele)
        B.append(row)

    # Matrix multiplication

    result=[]
    for i in range(rowsA):
        row=[]
        for j in range(colsB):
            sum=0
            for k in range(colsA):
                sum+=A[i][k]*B[k][j]
            row.append(sum)
        result.append(row)

    print("\nResultant Matrix (A x B)")
    for row in result:
        print(' '.join(str(x) for x in row))
```

```python
# Matrix Multiplication
rowsA = int(input("Enter rows for Matrix A: "))
colsA = int(input("Enter columns for Matrix A: "))
rowsB = int(input("Enter rows for Matrix B: "))
colsB = int(input("Enter columns for Matrix B: "))

if colsA != rowsB:
    print("Matrix multiplication not possible.")
else:
    A = [[int(input(f"A[{i+1}][{j+1}]: ")) for j in range(colsA)] for i in range(rowsA)]
    B = [[int(input(f"B[{i+1}][{j+1}]: ")) for j in range(colsB)] for i in range(rowsB)]

    result = [[sum(A[i][k]*B[k][j] for k in range(colsA)) for j in range(colsB)] for i in range(rowsA)]

    print("\nResultant Matrix (A × B):")
    for row in result:
        print(row)
```

```python
# Check if a square matrix is symmetric (A == A^T)

n=int(input("Enter size of matrix (n x n):"))

if n==0:
    print("Empty Matrix.")
else:
    mat=[]
    for i in range(n):
        row=[]
        for j in range(n):
            ele=int(input(f"Enter element at position ({i+1},{j+1})"))
            row.append(ele)
        mat.append(row)
    
    is_symmetric=True
    for i in range(n):
        for j in range(n):
            if mat[i][j]!=mat[j][i]:
                is_symmetric=False
                break
    print("\nMatrix: ")
    for row in mat:
        print(' '.join(str(x) for x in row))

    if is_symmetric:
        print("\nMatrix is Symmetric")
    else:
        print("\nMatrix is not Symmetric")
```

```python
#sum of primary secondary diagonals
n=int(input("Enter size of matrix (n x n)"))

if n==0:
    print("Empty Matrix")
else:
    mat=[]
    primary_sum=0
    secondary_sum=0

    for i in range(n):
        row=[]
        for j in range(n):
            ele=int(input(f"Enter element for position ({i+1},{j+1})"))
            row.append(ele)
            if i==j:
                primary_sum+=ele
            if i+j==n-1:
                secondary_sum+=ele
        mat.append(row)

    print("\nMatrix: ")
    for row in mat:
        print(' '.join(str(x) for x in row))

    print("\nSum of primary diagonals: ",primary_sum)
    print("Sum of secondary diagonals: ",secondary_sum)
```

```python
# Matrix Addition (User Input)
rows=int(input("Enter no of rows: "))
cols=int(input("Enter no of cols: "))

print("\nEnter elements for matrix A ")
A=[]
for i in range(rows):
    row=[]
    for j in range(cols):
        ele=int(input(f"Enter elements for position ({i+1},{j+1}): "))
        row.append(ele)
    A.append(row)

print("\nEnter elements for matrix B ")
B=[]
for i in range(rows):
    row=[]
    for j in range(cols):
        ele=int(input(f"Enter elements for position ({i+1},{j+1}): "))
        row.append(ele)
    B.append(row)


result=[]
for i in range(rows):
    row=[]
    for j in range(cols):
        row.append(A[i][j]+B[i][j])
    result.append(row)

print("\nResultant Matrix (A + B): ")
for row in result:
    print(' '.join(str(x) for x in row))
```

```python
# Sum of Each Row and Column in a Matrix
rows=int(input("Enter no of rows: "))
cols=int(input("Enter no of cols: "))

print("\nEnter elements for matrix ")
matrix=[]
for i in range(rows):
    row=[]
    for j in range(cols):
        ele=int(input(f"Enter elements for position ({i+1},{j+1}): "))
        row.append(ele)
    matrix.append(row)

print("\nMatrix: ")
for row in matrix:
    print(' '.join(str(x) for x in row))

print("\nSum of rows ")
for i in range(rows):
    sum_rows=sum(matrix[i])
    print(f"Sum of {i+1} row : {sum_rows}")

print("\nSum of columns ")
for j in range(cols):
    sum_cols=0
    for i in range(rows):
        sum_cols+=matrix[i][j]
    print(f"Sum of {j+1} column: {sum_cols}")
```

```python
# Display Upper and Lower Triangular Matrices
n=int(input("Enter the size of matrices: "))

matrix=[]
for i in range(n):
    row=[]
    for j in range(n):
        ele=int(input(f"Enter elements for position ({i+1},{j+1}): "))
        row.append(ele)
    matrix.append(row)

print("\n Original Matrix: ")
for row in matrix:
    print(row)                    #or   #print(' '.join(str(x) for x in row))

print("Upper Triangular Matrix: ")
for i in range(n):
    for j in range(n):
        if i>j:
            print("0",end=' ')
        else:
            print(matrix[i][j],end=' ')
    print()

print("Lower Triangular Matrix: ")
for i in range(n):
    for j in range(n):
        if i<j:
            print("0",end=' ')
        else:
            print(matrix[i][j],end=' ')
    print()
```

```python
# Check if matrix is identity matrix
n=int(input("Enter size of matrix (n x n): "))

matrix=[]
for i in range(n):
    row=[]
    for j in range(n):
        ele=int(input(f"Enter elements for position ({i+1},{j+1}): "))
        row.append(ele)
    matrix.append(row)

print("\nMatrix: ")
for row in matrix:
    print(' '.join(str(x) for x in row))

is_identity=True
for i in range(n):
    for j in range(n):
        if (i==j and matrix[i][j]!=1) or (i!=j and matrix[i][j]!=0):
            is_identity=False
            break

if is_identity:
    print("Matrix is an Identity Matrix.")
else:
    print("Matrix is not an Identity Matrix")
```

```python
# Check if matrix is upper triangular

n = int(input("Enter size of square matrix: "))
matrix = [[int(input(f"Enter element ({i+1},{j+1}): ")) for j in range(n)] for i in range(n)]

is_upper = True
for i in range(n):
    for j in range(n):
        if i > j and matrix[i][j] != 0:
            is_upper = False
            break

print("\nMatrix:")
for row in matrix:
    print(' '.join(str(x) for x in row))

if is_upper:
    print("\nMatrix is Upper Triangular ✅")
else:
    print("\nMatrix is NOT Upper Triangular ❌")
```

```python
# Count zeros in a matrix

rows=int(input("Enter no of rows: "))
cols=int(input("Enter no of cols: "))

matrix=[[int(input(f"Enter elements for position ({i+1},{j+1}):")) for j in range(cols)]for i in range(rows)]

print("\nMatrix:")
for row in matrix:
    print(' '.join(str(x) for x in row))

zero_count=0
for i in range(rows):
    for j in range(cols):
        if matrix[i][j]==0:
            zero_count+=1

print("\nNumber of zero elements in Matrix: ",zero_count)
```

```python
# Check if matrix is sparse
r=int(input("Enter no of rows: "))
c=int(input("Enter no of columns: "))

if r==0 and c==0:
    print("It is an Empty Matrix")

else:
    matrix=[[int(input(f"Enter elements for position ({i+1},{j+1}):")) for j in range(c)]for i in range(r)]
    
    total=r*c
    zero_count=0
    for i in range(r):
        for j in range(c):
            if matrix[i][j]==0:
                zero_count+=1
    
    print("\nOriginal Matrix: ")
    for row in matrix:
        print(' '.join(str(x) for x in row))
    
    if zero_count>total/2:
        print("It is a Sparse Matrix")
    else:
        print("It is not a Sparse Matrix")
```

```python
# Determinant of 3x3 matrix

print("Enter elements of 3x3 matrix:")
mat = [[int(input(f"({i+1},{j+1}): ")) for j in range(3)] for i in range(3)]

a, b, c = mat[0]
d, e, f = mat[1]
g, h, i = mat[2]

det = (a*(e*i - f*h)) - (b*(d*i - f*g)) + (c*(d*h - e*g))

print("\nMatrix:")
for row in mat:
    print(' '.join(str(x) for x in row))

print("\nDeterminant =", det)
```

```python
# Interchange two rows
rows=int(input("Enter no of rows: "))
cols=int(input("Enter no of cols: "))

matrix=[[int(input(f"Enter element for matrix ({i+1},{j+1}): ")) for j in range(cols)]for i in range(rows)]

r1=int(input("\nEnter first row number to swap: "))-1
r2=int(input("Enter second row number to swap: "))-1

print("\nOriginal Matrix: ")
for row in matrix:
    print(' '.join(str(x) for x in row))

matrix[r1],matrix[r2]=matrix[r2],matrix[r1]

print(f"\nMatrix after swapping matrix row {r1+1} and {r2+1}:")
for row in matrix:
    print(' '.join(str(x) for x in row))
```
