```python
class Dog:
    def __init__(self,name,breed,age):
        self.name=name           #public attribute
        self._breed=breed        #protected attribute 
        self.__age=age            #private attribute
    #public method
    def get_info(self):
        return f"Name: {self.name}, Breed: {self._breed},Age:{self.__age}"

    #getter and setter for private atribute
    def get_age(self):
        return self.__age

    def set_age(self,age):
        if age>0:
            self.__age=age
        else:
            print("Invalid age!")

#example usage
dog=Dog("Buddy","Labrador",3)

#accessing public member
print(dog.name)                  #accessible

#accessing protected member
print(dog._breed)                #accessible but discourged outside the class

#accessing private member using getter
print(dog.get_age())

#modifying private member using setter
dog.set_age(5)
print(dog.get_info())
```

```python
#Encapsulation Example: Car Speed Control
class Car:
    def __init__(self):
        self.__speed=0

    def increase_speed(self,value):
        if value>0:
            self.__speed+=value
            return f"Speed increased by {value}"
        else:
            return "Invalid speed value"

    def decrease_speed(self,value):
        if value>0 and self.__speed-value>=0:
            self.__speed-=value
            return f"Speed decreased by {value}"
        else:
            return "Speed can't be negative"

    def get_speed(self):
        return self.__speed

c1=Car()

print(c1.increase_speed(60))
print(c1.get_speed())

print(c1.decrease_speed(20))
print(c1.get_speed())

print(c1.decrease_speed(50))
print(c1.get_speed())
```

```python
class ATM:
    def __init__(self,pin,balance):
        self.__pin=pin
        self.__balance=balance
        
    def check_pin(self,entered_pin):
        return self.__pin==entered_pin

    def withdraw(self,entered_pin,amount):
        if self.__pin==entered_pin:
            if amount<=self.__balance:
                self.__balance -= amount
                return "Withdraw Sucessfull"
            else:
                return "Insufficient Balance"
        else:
            return "Invalid Pin!"

    def check_balance(self,entered_pin):
        if self.check_pin(entered_pin):
            return self.__balance
        else:
            return "Invalid Pin!"

atm=ATM(1234,1000)
print(atm.withdraw(1234,500))
print(atm.check_balance(123))
print(atm.check_balance(1234))
```

```python
#accessing parent class member in a child class
class base:
    def __init__(self,a):
        self.a=a

class derived(base):
    def __init__(self,x,y):
        base.a=x
        self.b=y
    def print_sum(self):
        self.s=base.a+self.b
        print("sum:",self.s)

d=derived(1,2)
d.print_sum()
```

```python
class Polygon:
    def __init__(self, sides):
        self.sides = sides

class Triangle(Polygon):
    def findArea(self):
        a, b, c = self.sides
        s = (a + b + c) / 2
        area = (s * (s - a) * (s - b) * (s - c)) ** 0.5
        print("Area of Triangle:", round(area, 2))

sides = [float(input(f"Enter side {i+1}: ")) for i in range(3)]

t = Triangle(sides)
t.findArea()
```

```python
#1. define two base classes containing single numeric attribute in each class. inherit these two classes to find larger attribute
class base1:
    def __init__(self,a):
        self.n1=a
        print("base1 class: n1 =",self.n1)

class base2:
    def __init__(self,b):
        self.n2=b
        print("base2 class: n2 =",self.n2)

class derived(base1,base2):
    def __init__(self,a,b):
        base1.__init__(self,a)
        base2.__init__(self,b)

    def find_larger(self):
        if self.n1>self.n2:
            l=self.n1
        else:
            l=self.n2
        print("Derived class: larger is",l)

d=derived(15,6)
d.find_larger()
```

```python
#2. define three base classes containing single numeric attribute in each class. 
# inherit these three classes to find largest attribute
class base1:
    def __init__(self,a):
        self.n1=a
        print("base1 class: n1 =",self.n1)

class base2:
    def __init__(self,b):
        self.n2=b
        print("base2 class: n2 =",self.n2)

class base3:
    def __init__(self,c):
        self.n3=c
        print("base3 class: n3 =",self.n3)

class derived(base1,base2,base3):
    def __init__(self,a,b,c):
        base1.__init__(self,a)
        base2.__init__(self,b)
        base3.__init__(self,c)

    def find_larger(self):
        if self.n1>self.n2 and self.n1>self.n3:
            l=self.n1
        elif self.n2>self.n1 and self.n2>self.n3:
            l=self.n2
        else:
            l=self.n3
        print("Derived class => Largest number is ",l)

d=derived(15,6,22)
d.find_larger()
```

2. Employee + Salary + Tax Calculation

Concept: Method reuse across parent classes

```python
class Employee:
    def __init__(self,emp_id,name):
        self.emp_id=emp_id
        self.name=name

    def show_emp(self):
        print("ID: ",self.emp_id)
        print("Name: ",self.name)

class Salary:
    def __init__(self,basic):
        self.basic=basic

    def calculate_salary(self):
        return self.basic + (0.3*self.basic)

class Tax:
    def calculate_tax(self,amount):
        return amount*0.1

class Payroll(Employee,Tax,Salary):
    def __init__(self,emp_id,name,basic):
        Employee.__init__(self,emp_id,name)
        Salary.__init__(self,basic)

    def show_payroll(self):
        salary=self.calculate_salary()
        tax=self.calculate_tax(salary)
        self.show_emp()
        print("Gross Salary: ",salary)
        print("Tax: ",tax)
        print("Net Salary: ",salary-tax)

p=Payroll(1,'Zoro',50000)
p.show_payroll()
```

```python
class Parent1():
    def show(self):
        print("Inside Parent1")

class Parent2():
    def display(self):
        print("Inside Parent2")

class Child(Parent1,Parent2):
    def show(self):
        print("Inside Child")

obj=Child()
obj.show()
obj.display()
```

```python
#multilevel inheritances
class Parent():
    def display(self):
        print("Inside Parent")

class Child(Parent):
    def show(self):
        print("Inside Child")

class GrandChild(Child):
    def show(self):
        print("Inside Grand Child")

g=GrandChild()
g.show()
g.display()
```

```python
class Printer:
    def print_data(self, data):
        if isinstance(data, int):
            print("Integer:", data)
        elif isinstance(data, str):
            print("String:", data)
        elif isinstance(data, list):
            print("List:", data)
        else:
            print("Unknown type")

p = Printer()
p.print_data(10)
p.print_data("Hello")
p.print_data([1, 2, 3])
```

âœ… Example: Overloading using *args (BEST Pythonic Way)

```python
class Calculator:
    def add(self, *args):
        total = 0
        for i in args:
            total += i
        return total

calc = Calculator()
print(calc.add(1, 2))
print(calc.add(1, 2, 3, 4))
print(calc.add(10))
```

## Data Abstraction

```python
from abc import ABC,abstractmethod

class Car(ABC):
    def __init__(self,brand,model,year):
        self.brand=brand
        self.model=model
        self.year=year

    @abstractmethod
    def printDetails(self):
        pass
    def accelerate(self):
        print("Speed up...")
    def break_applied(self):
        print("car Stopped")

class Hatchback(Car):
    def printDetails(self):
        print("Brand:",self.brand)
        print("Model:",self.model)
        print("Year:",self.year)
    def sunroof(self):
        print("Not having this feature")

class Suv(Car):
    def printDetails(self):
        print("Brand:",self.brand)
        print("Model:",self.model)
        print("Year:",self.year)
    def sunroof(self):
        print("Available")

car1=Hatchback("Maruti","Alto","2022")
car1.printDetails()
car1.accelerate()
car1.sunroof()
```

```python
from abc import ABC, abstractmethod

class PaymentGateway(ABC):
    def __init__(self,amount):
        self.amount=amount

    @abstractmethod
    def authenticate(self):
        pass

    @abstractmethod
    def make_payment(self):
        pass

    def payment_status(self):
        return "Payment Processing..."

class CreditCardPayment(PaymentGateway):
    def authenticate(self):
        print("Credit card authentication successfull")

    def make_payment(self):
        print(f"Paid {self.amount} using OTP")

class UPIPayment(PaymentGateway):
    def authenticate(self):
        print("UPI ID successfull")

    def make_payment(self):
        print(f"Paid {self.amount} using UPI code")

p=UPIPayment(2000)
p.authenticate()
p.make_payment()
print(p.payment_status())

print()

q=CreditCardPayment(10000)
q.authenticate()
q.make_payment()
print(q.payment_status())
```

File Handling System (Template Method Pattern)

```python
from abc import ABC,abstractmethod

class Filehandler:
    def process(self):
        self.open_file()
        self.read_file()
        self.close_file()

    @abstractmethod
    def open_file(self):
        pass

    @abstractmethod
    def read_file(self):
        pass

    def close_file(self):
        print("File closed safely")

class TextFile(Filehandler):
    def open_file(self):
        print("Text file opened")

    def read_file(self):
        print("Reading text file")

t=TextFile()
t.process()
```

Example 1: Banking System (Most Important Practice)
Concepts used

Encapsulation â†’ balance is private

Abstraction â†’ Account defines what operations exist

Inheritance â†’ SavingsAccount, CurrentAccount

```python
from abc import ABC,abstractmethod

class Account(ABC):
    def __init__(self,acc_no,name,balance):
        self.acc_no=acc_no
        self.name=name
        self.__balance=balance            #encapsulation
        
    def get_balance(self):
        return self.__balance

    def update_balance(self,amount):
        self.__balance+=amount

    @abstractmethod
    def withdraw(self, amount):
        pass

#Child classes (Inheritance + Implementation)
class SavingsAccount(Account):
    def withdraw(self,amount):
        if amount<=self.get_balance():
            self.update_balance(-amount)
            print("Savings withdraw successfull")
        else:
            print("Savings Insufficient balance")

class CurrentAccount(Account):
    def withdraw(self,amount):
        limit=5000
        if amount<self.get_balance()+limit:
            self.update_balance(-amount)
            print("Current Withdrawl Successfull")
        else:
            print("Current limit exceeded")

#usage
# Create Savings Account
savings = SavingsAccount(101, "Zoro", 20000)

print("Initial Balance:", savings.get_balance())

# Withdraw money
savings.withdraw(5000)

print("Balance after withdrawal:", savings.get_balance())


# Create Current Account
current = CurrentAccount(102, "Luffy", 10000)

print("\nInitial Balance:", current.get_balance())

# Withdraw within overdraft limit
current.withdraw(13000)

print("Balance after withdrawal:", current.get_balance())
```
