```python
#menu driven program to call function for
#a.interchange first half elements with second falf of a list
#b.count and sum even elements of list of numbers
#c.count and sum odd elements of list of numbers
#d.sum of all elements
#e.product of all elements

def swaphalf(list):
    length=len(list)
    if length%2==0:
        mid=length//2
        print("Mid(half value): ",mid)
        for i in range(mid):
            list[i],list[mid+i]=list[mid+i],list[i]
        print("List after interchanging is: ",list)
    else:
        print("Cannot half the list")

def count_sum_even(list):
    c=s=0
    for i in list:
        if i%2==0:
            c+=1
            s+=i
    print("Sum of %d even values in list is %d" %(c,s))

def count_sum_odd(list):
    c=s=0                         #c=count, s=sum
    for i in list:
        if i%2!=0:
            c+=1
            s+=i
    print("Sum of %d odd values is list is %d"%(c,s))

def sum_all(list):
    c=s=0
    for i in list:
        c+=1
        s+=i
    print(f"Sum of all {c} elements is {s}")

def product_all(list):
    c=0
    p=1
    for i in list:
        c+=1
        p*=i
    print(f"Product of all {c} elements is {p}")

print(
    '''***** Menu Driven Program *****
    1: Swap first half with second half
    2: Count and sum even elements
    3: Count and sum odd elements
    4: Sum of all elements
    5: Product of all elements
    --------------------------------'''
)
ch=int(input("\nSelect 1/2/3/4/5, any number to exit: "))
while(ch in(1,2,3,4,5)):
    list=eval(input("Enter your list: "))
    if ch==1:
        if len(list)%2==0:
            swaphalf(list)
        else:
            print("Size is odd")
    elif ch==2:
        count_sum_even(list)
    elif ch==3:
        count_sum_odd(list)
    elif ch==4:
        sum_all(list)
    elif ch==5:
        product_all(list)
    else:
        ch=int(input("\nSelect 1/2/3/4/5, any other number to exit: "))
    #or
    #ch=int(input("\nSelect 1/2/3/4/5, any other number to exit: "))
```

```python
#repeat 
def swap_half(list):
    length=len(list)
    if length%2==0:
        mid=length//2
        for i in range(mid):
            list[i],list[mid+i]=list[mid+i],list[i]
        print("List after interchanging elements: ",list)
    else:
        print("Elements of list are odd")

def sum_odd(list):
    c=s=0
    for i in list:
        if i%2!=0:
            c+=1
            s+=i
    print(f"Sum of {c} odd numbers in list is {s}")

def sum_even(list):
    c=s=0
    for i in list:
        if len(list)%2==0:
            c+=1
            s+=i
    print(f"Sum of {c} even numbers in list is {s}")

def sum_all(list):
    c=s=0
    for i in list:
        c+=1
        s+=i
    print(f"Sum of {c} numbers in list is {s}")

def product_all(list):
    c=0
    p=1
    for i in list:
        c+=1
        p*=i
    print(f"Product of {c} numbers in list is {p}")

print(
    '''--------- Select option for main menu -------------
    1. Swap half data
    2. Sum of odd elements
    3. Sum of even elements
    4. Sum of all elements
    5. Product of all elements'''
)
ch=int(input("Select option (1/2/3/4/5), or any other to exit: "))
while ch in (1,2,3,4,5):
    list=eval(input("Enter your list: "))
    if ch==1:
        if len(list)%2==0:
            swap_half(list)
        else:
            print("List is odd")
    elif ch==2:
        sum_odd(list)
    elif ch==3:
        sum_even(list)
    elif ch==4:
        sum_all(list)
    elif ch==5:
        product_all(list)
    ch=int(input("Select option (1/2/3/4/5), or any other to exit: "))
```

```python
#using recursion find GCD/HCF of two numbers
def gcd(m,n):
    if n==0:
        return m
    else:
        return gcd(n,m%n)

n1=int(input("Enter first no.: "))
n2=int(input("Enter second no.: "))
print('GCD is',gcd(n1,n2))
```

```python
# Lambda in Sorting Complex Tuples
# ðŸ‘‰ Task: Sort by second value descending
data = [(1, 3), (4, 1), (2, 5)]

data.sort(key=lambda x: x[1], reverse=True)
print(data)
```

```python
# Lambda with zip()
# ðŸ‘‰ Task: Multiply two lists
a = [1, 2, 3]
b = [4, 5, 6]

result = list(map(lambda x: x[0] * x[1], zip(a, b)))
print(result)
```

```python
# Lambda for Validation Logic
# ðŸ‘‰ Task: Password strength check
password = "Python@123"

is_strong = lambda p: (
    len(p) >= 8 and
    any(map(str.isupper, p)) and
    any(map(str.isdigit, p))
)

print(is_strong(password))
```
