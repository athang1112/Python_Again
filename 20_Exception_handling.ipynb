```python
def divide(divisor):
    for ele in divisor:
        try:
            print('36/',ele,end='=')
            print(36/ele)
        except(ZeroDivisionError):
            print('Error Division by Zero')
        except(TypeError):
            print('Error Non numeric Divisor')

divisor=[3,6,0,9,'6',12]
divide(divisor)
```

```python
class MarksValidation(Exception):
    def __init__(self,data):
        self.data=data
    def __str__(self):
        return repr(self.data)
marks=int(input("Enter marks obtained"))
try:
    if(marks>=500):
        raise MarksValidation("Invalid: must be less than 500")
except MarksValidation as mv:
    print("Received Error: ",mv.data)
else:
    print("Valid marks")
```

Division with exception handling

```python
try:
    num1=int(input("Enter first number: "))
    num2=int(input("Enter second number: "))
    result=num1/num2

except ValueError:
    print('Invalid input! Please enter number only')
except ZeroDivisionError:
    print('Cannot be divided by zero')
else:
    print('Result: ',result)
```

Custome exception: Age validation

```python
class InvalidAgeError(Exception):
    pass

try:
    age=int(input("Enter age:"))
    if age<18 or age >60:
        raise InvalidAgeError("Age must be between 18 to 60.")

except InvalidAgeError as e:
    print("Error:",e)

else:
    print("Valid age")
```

File handling exception

```python
try:
    filename=input("Enter file name: ")
    file = open(filename,'r')
    content=file.read()
    print(content)

except FileNotFoundError:
    print("File not found")

except PermissionError:
    print("You dont have permission to access this file")

finally:
    try:
        file.close()
    except:
        pass
```

Custom Exception - Password Validation 

```python
class WeakPasswordError(Exception):
    pass

try:
    password = input("Enter password: ")
    if len(password) < 8 or not any(char.isdigit() for char in password):
        raise WeakPasswordError("Password must be at least 8 characters and contain a digit.")
except WeakPasswordError as e:
    print("Error:", e)
else:
    print("Strong password.")
```

Bank Withdrawl Program

```python
class InsufficientBalanceError(Exception):
    pass

balance = 10000

try:
    amount = int(input("Enter withdrawal amount: "))
    if amount > balance:
        raise InsufficientBalanceError("Insufficient balance.")
    balance -= amount
except InsufficientBalanceError as e:
    print("Error:", e)
else:
    print("Withdrawal successful. Remaining balance:", balance)
```

Marks Validation (Advanced)

```python
class NegativeMarksError(Exception):
    pass

class ExceedMarksError(Exception):
    pass

try:
    marks = int(input("Enter marks (0-500): "))
    if marks < 0:
        raise NegativeMarksError("Marks cannot be negative.")
    if marks > 500:
        raise ExceedMarksError("Marks cannot exceed 500.")
except (NegativeMarksError, ExceedMarksError) as e:
    print("Error:", e)
else:
    print("Valid marks.")
```
