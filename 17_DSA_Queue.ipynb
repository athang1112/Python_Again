#### Hospital Queue Program (Python)

```python
def isempty(queue):
    if queue==[]:
        return True
    else:
        return False

def add_patient(queue,name):
    queue.append(name)
    print(name,"added to waiting list")

def call_patient(queue):
    if isempty(queue):
        return 'underflow'
    else:
        patient=queue[0]
        queue.remove(patient)
        return patient

def display(queue):
    if isempty(queue):
        print("No patients waiting")   
    else:
        print("Patients in waiting list:\n")
        for i in range(len(queue)):
            if i==0:
                print(queue[i]," <-- Next Patient")
            elif i==len(queue)-1:
                print(queue[i]," <-- Last Patient")
            else:
                print(queue[i])
queue=[]
print("Hospital Management System")
print("1. Add patient")
print("2. Call patient")
print("3. Display waiting list\n")

while True:
    ch=int(input("Select option (1-3): "))
    if ch==1:
        name=input("Enter patient name: ")
        add_patient(queue,name)
        print()
        
    elif ch==2:
        patient=call_patient(queue)
        if patient=='underflow':
            print("No patients to call ")
        else:
            print("Doctor is calling patient",patient)
        print()

    elif ch==3:
        display(queue)
        print()

    a=input("Do you wish to coninue (y/n):")
    if a.lower()=='y':
        continue
    else:
        break
    
```

```python
def isempty(queue):
    if queue==[]:
        return True
    else:
        return False

def place_order(queue,order):
    queue.append(order)
    print(order,"Order placed")

def prepare_order(queue):
    if isempty(queue):
        return 'underflow'
    else:
        order=queue[0]
        queue.remove(order)
        return order

def display_orders(queue):
    if isempty(queue):
        print("No pending food orders")
    else:
        print("Pending food orders:\n ")
        for i in range(len(queue)):
            if i==0:
                print(queue[i],"<-- Preparing Next")
            elif i==len(queue)-1:
                print(queue[i],"<-- Last order")
            else:
                print(queue[i])

queue=[]

print("Food order management and system")
print("1. Place order")
print("2. Prepare order")
print("3. Show pending orders")

while True:
    ch=int(input("Select option (1-3): "))
    if ch==1:
        order = input("Enter food item: ")
        place_order(queue,order)

    elif ch==2:
        order=prepare_order(queue)
        if order=='underflow':
            print("No order to prepare")
        else:
            print("Order prepared",order)
        print()

    elif ch==3:
        display_orders(queue)
        print()

    else:
        print("invalid choice!")

    a=input("Do you wich to continue (y/n): ")
    if a.lower()=='y':
        continue
    else:
        break
```
