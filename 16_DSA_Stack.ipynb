## Stack

```python
def isempty(stk):
    if stk==[]:
        return True
    else:
        return False

def pop(stk):
    if isempty(stk):
        return "underflow"
    else:
        item=stk.pop()
        if len(stk)==0:
            top=None
        else:
            top=len(stk)-1
        return item

def push(stk,item):
    stk.append(item)
    top=len(stk)-1

def peep(stk):
    if isempty(stk):
        return 'underflow'
    else:
        top=len(stk)-1
        return stk[top]

def display(stk):
    if isempty(stk):
        print("Stack Empty")
    else:
        top=len(stk)-1
        print(stk[top],'<-top')
        for a in range(top-1,-1,-1):
            print(stk[a])

stack=[]
top=None
print("Stack Operations")
print("1.Push")
print("2.Peep")
print("3.Pop")
print("4.Display")
print("5.Exit")

while True:
    ch=int(input("Select Option (1-5):"))
    if ch==1:
        item=int(input("Enter value:"))
        push(stack,item)
    elif ch==2:
        item=peep(stack)
        if item=='underflow':
            print("Underflow! Stack is empty!")
        else:
            print("Top of the stack is:",item)
    elif ch==3:
        item=pop(stack)
        if item=='underflow':
            print("Underflow!,stack is empty")
        else:
            print("Popped item is:",item)
    elif ch==4:
        display(stack)
    elif ch==5:
        break
    else:
        print("Invalid Option!")
    print()
```

```python
def isempty(stk):
    return stk==[]

def push(stk,transaction):
    stk.append(transaction)

def pop(stk):
    if isempty(stk):
        return "Underflow"
    return stk.pop()

def peep(stk):
    if isempty(stk):
        return "Underflow"
    return stk[-1]

def display(stk):
    if isempty(stk):
        print("No Transaction found")
    else:
        print("\nTransaction history (latest on top)")
        print(stk[-1],' <-top')
        for a in range(len(stk)-2,-1,-1):
            print(stk[a])

stack=[]

print("ATM Transaction Stack")
print("1. Add transaction")
print("2. View last transaction")
print("3. Cancel Last transaction")
print("4. Show all transaction")
print("5. Exit")

while True:
    ch=int(input("\nSelect option (1-5):"))

    if ch==1:
        t=input("Enter transaction (Deposit/Withdraw)")
        push(stack,t)

    elif ch==2:
        t=peep(stack)
        if t=='underflow':
            print("No transaction available")
        else:
            print("Last tranaction:",t)

    elif ch==3:
        t=pop(stack)
        if t=='Underflow':
            print("No transaction available")
        else:
            print("Cancelled tranaction:",t)

    elif ch==4:
        display(stack)

    elif ch==5:
        print("Thank you")
        break

    else:
        print("invalid choice")
```

```python
def isempty(stk):
    return stk == []

def visit_page(stk,page):
    stk.append(page)
    print(page, "visited")

def current_page(stk):
    if isempty(stk):
        return 'underflow'
    return stk[-1]

def go_back(stk):
    if isempty(stk):
        return 'underflow'
    return stk.pop()

def display_history(stk):
    if isempty(stk):
        print("No browsing history")
    else:
        print("Browsing history (Current page)")
        print(stk[-1]," <-- Current page")
        for a in range(len(stk)-2,-1,-1):
            print(stk[a])

stack=[]

print("Browser History Stack")
print("1. Visit Page")
print("2. View Current Page")
print("3. Go Back")
print("4. Show History")
print("5. Exit")

while True:
    ch = int(input("\nSelect option (1-5): "))
    if ch==1:
        page=input("Enter page name:")
        visit_page(stack,page)

    elif ch==2:
        page=current_page(stack)
        if page=='underflow':
            print("No page available")
        else:
            print("Current Page",page)

    elif ch==3:
        page=go_back(stack)
        if page=='underflow':
            print("No page to go back")
        else:
            print("Went back from",page)

    elif ch == 4:
        display_history(stack)

    elif ch == 5:
        print("Thank you")
        break

    else:
        print("Invalid choice")
```
