Dry Run (Step by Step)

First iteration:

digit = 1234 % 10 = 4

rev = 0 * 10 + 4 = 4

n = 1234 // 10 = 123

Second iteration:

digit = 123 % 10 = 3

rev = 4 * 10 + 3 = 43

Third iteration:

digit = 12 % 10 = 2

rev = 43 * 10 + 2 = 432

Fourth iteration:

digit = 1 % 10 = 1

rev = 432 * 10 + 1 = 4321

n = 1 // 10 = 0

```python
#prime number checking
num=int(input("Enter number to check: "))
if num>1:
    for i in range(2,num):
        if num%i==0:
            print(num,"is not a prime number")
            break
    else:
        print(num," is a prime number")
else:
    print(num,"is not a prime number")
```

```python
lower_value=int(input("Enter lower bounding value "))
upper_value=int(input("Enter upper bounding value "))
print(f"Prime numbers between {lower_value} and {upper_value}:")
for n in range(lower_value,upper_value+1):
    if n>1:
        for i in range(2,n):
            if n%i==0:
                break
        else:
            print(n)
        
```

```python
#palindrome number
n=int(input("Enter number to check palindrome:"))
temp=n
rev=0
while temp>0:
    rev=rev*10+temp%10
    temp//=10

if rev==n:
    print(n,"is a palindrome.")
else:
    print(n,"is not a palindrome.")
```

```python
#prime factors of number
n=int(input("Enter a number: "))
print(f"Prime factors of {n} are: ",end=" ")
i=2
while i<=n:
    if n%i==0:
        print(i,end=' ')
        n=n//i
    else:
        i+=1
```

```python
#factorial
num=int(input("Enter number to find factorial: "))
fact=1
for i in range(1,num+1):
    fact*=i
print(fact)
```

```python
num=int(input("Enter a number: "))
temp=num
sum=0
while temp>0:
    sum+=temp%10
    temp//=10
print(f"Sum of digits of {num} is {sum}")
```

```python
###### perfect numbers
num=int(input("Enter number to find perfect number: "))
sum=0
for i in range(1,num):
    if num%i==0:
        sum+=i
if sum==num:
    print(num,"is a perfect number")
else:
    print(num,"is not a perfect number")
```

```python
##number pyramid
rows = int(input("Enter number of rows: "))

for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

```python
###### reverse number
n=123
rev=0
while n>0:
    digit=n%10
    rev=rev*10+digit
    n//=10
print(rev)
```
