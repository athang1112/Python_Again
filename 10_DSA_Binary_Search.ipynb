üí†1.  Basic Search

```python
def binarySearch(arr,target,low,high):
    while low<=high:
        mid=low+(high-low)//2

        if arr[mid]==target:
            return mid
            
        if arr[mid]<target:
            low=mid+1
        else:
            high=mid-1
    else:
        print(target,"Not found in list")

arr=[2,3,4,7,9,10,11]
n=len(arr)
target=11
low=0
high=n
index=binarySearch(arr,target,low,high)
print(index)
```

üí†3. Find First Occurrence

```python
def first_occurrence(arr, target):
    low, high = 0, len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            # check if it is the first occurrence
            if mid == 0 or arr[mid - 1] != target:
                return mid
            else:
                high = mid - 1   # move left
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1
arr = [1, 2, 2, 2, 3, 4, 5]
first_occurrence(arr, 2)  # returns 1
```

```python
def first_occurrence(arr, target):
    low, high = 0, len(arr) - 1
    result = -1
    while low <= high:
        mid = (low + high)//2
        if arr[mid] == target:
            result = mid
            high = mid - 1  # move left to find first
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return result

arr = [1, 2, 2, 2, 3, 4, 5]
first_occurrence(arr, 2)  # returns 1
```

When we find target, we don‚Äôt stop.

We move high = mid - 1 to continue searching left side.

üí†4. Find Last Occurrence

```python
def last_occurrence(arr, target):
    low, high = 0, len(arr) - 1
    result = -1
    while low <= high:
        mid = (low + high)//2
        if arr[mid] == target:
            result = mid
            low = mid + 1  # move right to find last
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return result
arr = [1, 2, 2, 2, 3, 4, 5]
target = 2
last_occurrence(arr, 2)  # returns 3
```

Similar to first occurrence, but now we move right to find the last index of the target.

üí†5.  Smallest Number ‚â• Target (Lower Bound)

```python
def lower_bound(arr, target):
    low, high = 0, len(arr) - 1
    result = -1
    while low <= high:
        mid = (low + high)//2
        if arr[mid] >= target:
            result = mid
            high = mid - 1  # go left to find smaller ‚â• target
        else:
            low = mid + 1
    return result
arr = [1, 3, 5, 7, 9]
lower_bound(arr, 6)  # returns 3 (element 7)
```

Find the smallest number that is ‚â• target.

Keep updating result when arr[mid] >= target.

Move left to find a smaller valid number.

üí†6. Largest Number ‚â§ Target (Upper Bound)

```python
def upper_bound(arr, target):
    low, high = 0, len(arr) - 1
    result = -1
    while low <= high:
        mid = (low + high)//2
        if arr[mid] <= target:
            result = mid
            low = mid + 1  # go right to find bigger ‚â§ target
        else:
            high = mid - 1
    return result
arr = [1, 3, 5, 7, 9]
target = 6

index = upper_bound(arr, target)
if index != -1:
    print("Largest number <= target:", arr[index], "at index", index)
else:
    print("No number <= target found")
```

Find the largest number that is ‚â§ target.

Update result when arr[mid] <= target.

Move right to see if there is a bigger valid number.

üí†7. Search in Rotated Sorted Array

```python
#In a rotated array, one half is always sorted
def search_rotated(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high)//2
        if arr[mid] == target:
            return mid
        if arr[low] <= arr[mid]:  # left sorted
            if arr[low] <= target < arr[mid]:
                high = mid - 1
            else:
                low = mid + 1
        else:  # right sorted
            if arr[mid] < target <= arr[high]:
                low = mid + 1
            else:
                high = mid - 1
    return -1
arr = [10, 12, 15, 2, 5, 6, 8]
search_rotated(arr, 5)  # returns 4


arr = [10, 12, 15, 2, 5, 6, 8]
target = 5

index = search_rotated(arr, target)
if index != -1:
    print("Target found at index", index)
else:
    print("Target not found")
```

| Iteration | low | high | mid | arr[mid] | Sorted half       | Target in sorted half? | Action   |
| --------- | --- | ---- | --- | -------- | ----------------- | ---------------------- | -------- |
| 1         | 0   | 6    | 3   | 2        | Right `[2,5,6,8]` | Yes                    | low = 4  |
| 2         | 4   | 6    | 5   | 6        | Left `[5,6]`      | Yes                    | high = 4 |
| 3         | 4   | 4    | 4   | 5        | Only one element  | ‚úÖ Found                | return 4 |


What we did:

Identify which half of the array is sorted.

If target is in the sorted half ‚Üí move there.

Else ‚Üí move to the other half.

Repeat until found.

üí†8. Peak Element in Mountain Array

```python
def peak_element(arr):
    low, high = 0, len(arr) - 1
    while low < high:
        mid = (low + high)//2
        if arr[mid] < arr[mid + 1]:
            low = mid + 1
        else:
            high = mid
    return low

arr = [1, 3, 5, 7, 6, 4, 2]

index = peak_element(arr)
print("Peak element:", arr[index], "at index", index)
```

What we did:

Find element that is greater than neighbors.

If arr[mid] < arr[mid+1] ‚Üí peak is right

Else ‚Üí peak is mid or left

Stop when low == high ‚Üí peak index.

üí†9. Smallest Element in Rotated Array

```python
def min_in_rotated(arr):
    low, high = 0, len(arr) - 1
    while low < high:
        mid = (low + high)//2
        if arr[mid] > arr[high]:
            low = mid + 1
        else:
            high = mid
    return arr[low]

arr = [4, 5, 6, 7, 0, 1, 2]

minimum = min_in_rotated(arr)
print("Minimum element in rotated array:", minimum)
```

What we did:

Find the minimum element in rotated array.

Compare arr[mid] with arr[high] to determine which side has min.

Keep moving pointers until low == high.

üí†10.  Count Occurrences of a Number

```python
def count_occurrences(arr, target):
    def first_occurrence(arr, target):
        low, high = 0, len(arr) - 1
        result = -1
        while low <= high:
            mid = (low + high)//2
            if arr[mid] == target:
                result = mid
                high = mid - 1
            elif arr[mid] < target:
                low = mid + 1
            else:
                high = mid - 1
        return result

    def last_occurrence(arr, target):
        low, high = 0, len(arr) - 1
        result = -1
        while low <= high:
            mid = (low + high)//2
            if arr[mid] == target:
                result = mid
                low = mid + 1
            elif arr[mid] < target:
                low = mid + 1
            else:
                high = mid - 1
        return result

    first = first_occurrence(arr, target)
    last = last_occurrence(arr, target)
    if first == -1:
        return 0
    return last - first + 1


arr = [1, 2, 2, 2, 3, 4, 5]
target = 2

count = count_occurrences(arr, target)
print("Target occurs", count, "times")
```

What we did:

Use first and last occurrence functions.

Count total occurrences = last - first + 1

If target not found ‚Üí return 0
