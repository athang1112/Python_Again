```python
#printing list of tuple in table format(method1)
list1=[('Naruto','Leaf',92.34),
       ('Luffy','East Blue',85.85),
       ('Ichigo','Karakura Town',82.12)]
print("Name\t\tCity\t\tScore")
for a in range(40):
    print('-',end='')
print()
for ele1,ele2,ele3 in list1:
    print("{:16}{:16}{}".format(ele1,ele2,ele3))
```

```python
#printing list of tuple in table format(method2)
list2=[('Naruto','Leaf',92.34),
       ('Luffy','East Blue',85.85),
       ('Ichigo','Karakura Town',82.12)]
column_width=16
print("Name\t\tCity\t\tScore")
for a in range(40):
    print('-',end='')
print()
for tup in list2:
    for ele in tup:
        print(str(ele).ljust(column_width),end='')
    print()
```

```python
#check if any list element exist in tuple
tup=(51,35,12,45,73)
print("The tuple elements are "+str(tup))
lst=[12,0,4,35,23]
print("The list elements are"+str(lst))
status=False
for ele in lst:
    if ele in tup:
        status=True
        break
print("If any element in list exist in tuple? \n"+str(status))
```

```python
#convert tuple to string (method1)
tup=('B', 'a', 'n', 'k', 'a', 'i')
print("Tuple items are",tup)
str1=''.join(tup)
print("String from tuple is",str1)
```

```python
#convert tuple to string (method2)
tup=('B', 'a', 'n', 'k', 'a', 'i')
print("Tuple items are",tup)
str2=''
for t in tup:
    str2=str2+t
print("String from tuple is",str2)
```

```python
#convert tuple to string (method3)
tup=('B', 'a', 'n', 'k', 'a', 'i')
print("Tuple items are",tup)
str3=''
for i in range(len(tup)):
    str3=str3+tup[i]
print("String from tuple is",str3)
```
