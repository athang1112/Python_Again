## Selection Sort

```python
def selection(l):
    i=0
    while i<len(l):
        smallest=min(l[i:])
        ind=l.index(smallest)
        l[i],l[ind]=l[ind],l[i]
        i+=1
    print(l)

l=eval(input("Enter list: "))
selection(l)
```

```python
#using for loops (standard approach)
def selection(l):
    for i in range(len(l)):
        min_i=i
        for j in range(i+1,len(l)):
            if l[j]<l[min_i]:
                min_i=j
        l[i],l[min_i]=l[min_i],l[i]
    print(l)

selection([2,8,6,4,9,1,3])
```

```python
#using while loops
def selection(l):
    i=0
    while i<len(l)-1:
        min_i=i
        j=i+1
        while j<len(l):
            if l[j]<l[min_i]:
                min_i=j
            j+=1
        l[i],l[min_i]=l[min_i],l[i]
        i+=1
    print(l)

selection([2,8,6,4,9,1,3])
```

```python
#by tracking smallest value and its position
def selection(l):
    for i in range(len(l)-1):
        smallest=l[i]
        pos=i
        for j in range(i+1,len(l)):
            if l[j]<smallest:
                smallest=l[j]
                pos=j
        l[i],l[pos]=l[pos],l[i]
    print(l)

selection([8, 6,3, 4, 2])
```

```python
#starting inner loop from current index
def selection(l):
    i = 0
    while i < len(l):
        min_i = i
        for j in range(i, len(l)):
            if l[j] < l[min_i]:
                min_i = j
        l[i], l[min_i] = l[min_i], l[i]
        i += 1
    print(l)

selection([10, 1, 9, 2])
```

```python
#Sort using slicing and min()
def selection(l):
    for i in range(len(l)):
        m = min(l[i:])
        idx = l.index(m, i)
        l[i], l[idx] = l[idx], l[i]
    print(l)

selection([6, 2, 9, 3, 1])
```

```python
#Early Stop Version (FIXED â—)
#ðŸ“Œ Note:
# Selection sort cannot truly early-stop safely like bubble sort.
# This version is logically correct but still does all passes.

def selection_early_stop(l):
    i = 0
    while i < len(l) - 1:
        min_i = i
        j = i + 1
        while j < len(l):
            if l[j] < l[min_i]:
                min_i = j
            j += 1
        if min_i != i:
            l[i], l[min_i] = l[min_i], l[i]
        i += 1
    print(l)

selection_early_stop([5,1,8,3,2,4])
```

```python
#Sort with Duplicates
def selection_duplicates(l):
    i = 0
    while i < len(l) - 1:
        min_i = i
        j = i + 1
        while j < len(l):
            if l[j] < l[min_i]:
                min_i = j
            j += 1
        l[i], l[min_i] = l[min_i], l[i]
        i += 1
    print(l)

selection_duplicates([4, 2, 2, 5, 1])
```

```python
#Sort on First k Elements
def selection_partial(l, k):
    i = 0
    while i < k - 1:
        min_i = i
        j = i + 1
        while j < k:
            if l[j] < l[min_i]:
                min_i = j
            j += 1
        l[i], l[min_i] = l[min_i], l[i]
        i += 1
    print(l)

selection_partial([9, 4, 6, 2, 8, 1], 5)
```

```python
#Sort with Negative Numbers

def selection_negative(l):
    i = 0
    while i < len(l) - 1:
        min_i = i
        j = i + 1
        while j < len(l):
            if l[j] < l[min_i]:
                min_i = j
            j += 1
        l[i], l[min_i] = l[min_i], l[i]
        i += 1
    print(l)

selection_negative([-3, 1, -7, 4, 0])
```

```python
#Sort with Index Tracking
def selection_index_track(l):
    i = 0
    while i < len(l) - 1:
        min_i = i
        j = i + 1
        while j < len(l):
            if l[j] < l[min_i]:
                min_i = j
            j += 1
        print("pass", i + 1, "-> index", min_i)
        l[i], l[min_i] = l[min_i], l[i]
        i += 1
    print(l)

selection_index_track([8, 3, 6, 1])
```

```python
#Sort on Tuples   
def selection_tuple(l):
    i = 0
    while i < len(l) - 1:
        min_i = i
        j = i + 1
        while j < len(l):
            if l[j][1] < l[min_i][1]:
                min_i = j
            j += 1
        l[i], l[min_i] = l[min_i], l[i]
        i += 1
    print(l)
#(index/value , key_to_sort_by)
selection_tuple([(1, 4), (2, 1), (3, 3),(6,2)])
```

```python
#Stable Selection Sort
def stable_selection(l):
    i = 0
    while i < len(l) - 1:
        min_i = i
        j = i + 1
        while j < len(l):
            if l[j][0] < l[min_i][0]:
                min_i = j
            j += 1

        key = l[min_i]
        while min_i > i:
            l[min_i] = l[min_i - 1]
            min_i -= 1
        l[i] = key
        i += 1

    print(l)

stable_selection([(1,'a'), (1,'b'), (0,'c'),(0,'d')])
```

```python
def bubble(l):
    n = len(l)
    for i in range(n):
        for j in range(0, n - i - 1):
            if l[j] > l[j + 1]:
                l[j], l[j + 1] = l[j + 1], l[j]
    print("List after sorting is", l)


def insertion(l):
    for i in range(1, len(l)):
        key = l[i]
        j = i - 1
        while j >= 0 and key < l[j]:
            l[j + 1] = l[j]
            j -= 1
        l[j + 1] = key
    print("List after sorting is", l)


def selection(l):
    i = 0
    while i < len(l):
        smallest = min(l[i:])
        ind = l.index(smallest)
        l[i], l[ind] = l[ind], l[i]
        i += 1
    print("List after sorting is", l)


# -------- 4. Merge Sort --------
def merge_sort(l):
    if len(l) > 1:
        mid = len(l) // 2
        L = l[:mid]
        R = l[mid:]

        merge_sort(L)
        merge_sort(R)

        i = j = k = 0
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                l[k] = L[i]
                i += 1
            else:
                l[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            l[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            l[k] = R[j]
            j += 1
            k += 1


# -------- 5. Quick Sort --------
def quick_sort(l):
    if len(l) <= 1:
        return l
    pivot = l[len(l) // 2]
    left = [x for x in l if x < pivot]
    middle = [x for x in l if x == pivot]
    right = [x for x in l if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)


# -------- 6. Heap Sort --------
def heapify(l, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and l[left] > l[largest]:
        largest = left
    if right < n and l[right] > l[largest]:
        largest = right

    if largest != i:
        l[i], l[largest] = l[largest], l[i]
        heapify(l, n, largest)


def heap_sort(l):
    n = len(l)
    for i in range(n // 2 - 1, -1, -1):
        heapify(l, n, i)

    for i in range(n - 1, 0, -1):
        l[i], l[0] = l[0], l[i]
        heapify(l, i, 0)
    print("List after sorting is", l)


# -------- 7. Shell Sort --------
def shell_sort(l):
    gap = len(l) // 2
    while gap > 0:
        for i in range(gap, len(l)):
            temp = l[i]
            j = i
            while j >= gap and l[j - gap] > temp:
                l[j] = l[j - gap]
                j -= gap
            l[j] = temp
        gap //= 2
    print("List after sorting is", l)


print('''
1. Bubble Sort
2. Selection Sort
3. Insertion Sort
4. Merge Sort
5. Quick Sort
6. Heap Sort
7. Shell Sort
''')

while True:
    choice = int(input("Enter your choice: "))
    l = eval(input("Enter a list: "))

    if choice == 1:
        bubble(l)
    elif choice == 2:
        selection(l)
    elif choice == 3:
        insertion(l)
    elif choice == 4:
        merge_sort(l)
        print("List after sorting is", l)
    elif choice == 5:
        l = quick_sort(l)
        print("List after sorting is", l)
    elif choice == 6:
        heap_sort(l)
    elif choice == 7:
        shell_sort(l)
    else:
        print("Invalid input")

    a = input("Do you want to continue (y/n): ")
    if a.lower() != 'Yesyes':
        break
```
