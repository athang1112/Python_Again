1Ô∏è‚É£ Linear Search

```python
def search(arr, target):
    n=len(arr)
    for i in range(n):
        if arr[i]==target:
            return f"{target} found at position {i+1}"
            break
    else:
        print("Target not found in list")

arr=[2,3,4,7,1,5]
target=1
index=search(arr,target)
print(index)
```

2Ô∏è‚É£ Count Occurrences

```python
def count_occurences(arr,key):
    count=0
    for x in arr:
        if x==key:
            count+=1
    return count

arr=[2,4,6,8,3,2,4,2]
print(count_occurences(arr,2))
```

3Ô∏è‚É£ First and Last Occurrence 

```python
def first_last(arr,key):
    first=-1
    last=-1
    for i in range(len(arr)):
        if arr[i]==key:
            if first==-1:
                first=i
            last=i
    return first, last

arr=[2,6,5,4,7,8,3,4,5,3,6,3]
print(first_last(arr,3))
```

4Ô∏è Element Appearing Only Once

```python
def find_unique(arr):
    for i in range(len(arr)):
        count = 0
        for j in range(len(arr)):
            if arr[i] == arr[j]:
                count += 1
        if count == 1:
            return arr[i]

arr = [2, 3, 5, 4, 5, 3, 2]
print(find_unique(arr))
```

5Ô∏è‚É£ Find All Duplicates

```python
def find_duplicates(arr):
    duplicates = []
    for i in range(len(arr)):
        count = 0
        for j in range(len(arr)):
            if arr[i] == arr[j]:
                count += 1
        if count > 1 and arr[i] not in duplicates:
            duplicates.append(arr[i])
    return duplicates

arr = [1, 2, 3, 1, 3, 6, 6]
print(find_duplicates(arr))
```

6Ô∏è‚É£ Majority Element (> n/2)

```python
def majority_element(arr):
    n = len(arr)
    for i in range(n):
        count = 0
        for j in range(n):
            if arr[i] == arr[j]:
                count += 1
        if count > n // 2:
            return arr[i]
    return None

arr = [2, 2, 1, 2, 3, 2, 2]
print(majority_element(arr))
```

7Ô∏è‚É£ Min, Max, Second Min, Second Max

```python
def min_max_second(arr):
    arr = list(set(arr))
    arr.sort()
    return arr[0], arr[1], arr[-1], arr[-2]

arr = [10, 3, 5, 8, 2, 9]
print(min_max_second(arr))
```

8Ô∏è‚É£ Count Elements Greater Than Average

```python
def count_greater_than_avg(arr):
    avg = sum(arr) / len(arr)
    count = 0
    for x in arr:
        if x > avg:
            count += 1
    return count

arr = [2, 4, 6, 8]
print(count_greater_than_avg(arr))
```

9Ô∏è‚É£ Search in Rotated Array (Linear)

```python
def search_rotated(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

arr = [4, 5, 6, 1, 2, 3]
print(search_rotated(arr, 2))
```

üîü Equilibrium Index

```python
def equilibrium_index(arr):
    for i in range(len(arr)):
        if sum(arr[:i]) == sum(arr[i+1:]):
            return i
    return -1

arr = [1, 3, 5, 2, 2]
print(equilibrium_index(arr))
```

1Ô∏è‚É£1Ô∏è‚É£ Palindrome Array

```python
def is_palindrome(arr):
    start = 0
    end = len(arr) - 1
    while start < end:
        if arr[start] != arr[end]:
            return False
        start += 1
        end -= 1
    return True

arr = [1, 2, 3, 2, 1]
print(is_palindrome(arr))
```
