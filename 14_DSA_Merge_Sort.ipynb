## Merge Sort

```python
# Merge Sort Program (Recursive)
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2

        left = arr[:mid]
        right = arr[mid:]

        # divide
        merge_sort(left)
        merge_sort(right)

        i = j = k = 0

        # merge
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                arr[k] = left[i]
                i += 1
            else:
                arr[k] = right[j]
                j += 1
            k += 1

        # remaining elements
        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1


# driver code
a = [38, 27, 43, 3, 9, 82, 10]
merge_sort(a)
print("Sorted array:", a)
```

```python
def mergeSort(arr):
  if len(arr) <= 1:
    return arr

  mid = len(arr) // 2
  leftHalf = arr[:mid]
  rightHalf = arr[mid:]

  sortedLeft = mergeSort(leftHalf)
  sortedRight = mergeSort(rightHalf)

  return merge(sortedLeft, sortedRight)

def merge(left, right):
  result = []
  i = j = 0

  while i < len(left) and j < len(right):
    if left[i] < right[j]:
      result.append(left[i])
      i += 1
    else:
      result.append(right[j])
      j += 1

  result.extend(left[i:])
  result.extend(right[j:])

  return result

mylist = [3, 7, 6, -10, 15, 23.5, 55, -13]
mysortedlist = mergeSort(mylist)
print("Sorted array:", mysortedlist)
```

```python
def merge_sort_trace(arr, level=0):
    indent = "  " * level
    print(f"{indent}mergeSort called on {arr}")

    if len(arr) <= 1:
        print(f"{indent}return {arr}")
        return arr

    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]

    print(f"{indent}split into {left} and {right}")

    sorted_left = merge_sort_trace(left, level + 1)
    sorted_right = merge_sort_trace(right, level + 1)

    merged = merge_trace(sorted_left, sorted_right, level)
    print(f"{indent}merged result: {merged}")
    return merged


def merge_trace(left, right, level):
    indent = "  " * level
    print(f"{indent}merge {left} and {right}")

    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        print(f"{indent} compare {left[i]} and {right[j]}")
        if left[i] < right[j]:
            result.append(left[i])
            print(f"{indent}  take {left[i]} → {result}")
            i += 1
        else:
            result.append(right[j])
            print(f"{indent}  take {right[j]} → {result}")
            j += 1

    if i < len(left):
        print(f"{indent} append remaining left {left[i:]}")
    if j < len(right):
        print(f"{indent} append remaining right {right[j:]}")

    result.extend(left[i:])
    result.extend(right[j:])

    return result


arr = [5, 8, 3, 6, 2]
print("\nFINAL SORTED ARRAY:", merge_sort_trace(arr))
```
