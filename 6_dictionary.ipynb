```python
#declare directory to  create new, edit, and display multiple input for name and mobile number
d=dict()
while True:
    print(
    '''---- Main Menu ----
    1. Create Directory
    2. Edit Directory
    3. Display Directory
    ----------------------'''
    )

    o=int(input("Select Main Menu Option: "))
    if o==1:  #creating contacts
        n=int(input("Enter no of contacts: "))
        for i in range(n):
            key=input("Name: ")
            value=int(input("Mobile no.: "))
            d[key]=value
    elif o==2:  #editing contacts
        print(
            '''-------Edit Menu-------
            1. Add a new contact
            2. Change a contact number
            '''
        )
        s=int(input('Select edit menu options: '))
        if s==1:
            key=input("Enter new contact name: ")
            value=int(input('Enter mobile number: '))
            d[key]=value
        elif s==2:
            key=input("Enter existing contact name: ")
            value=int(input('Enter new mobile number: '))
            d[key]=value
        else:
            print("Invalid Input !!! ")
    elif o==3:  #display contacts
        print("Name (Mobile number)\n")
        for key in d:
            print(key,'(',d[key],")")
    else:
        print("Invalid Input !!!")
    q=input("Do you wish to continue? (y/n):")
    if q in 'nN':
        break
```

```python
#Nested Dictionary Aggregation (Advanced)
#Calculate total marks and percentage for each student.
students = {
    "Amit": {"Math": 85, "Science": 90, "English": 78},
    "Riya": {"Math": 92, "Science": 88, "English": 95},
    "Karan": {"Math": 70, "Science": 65, "English": 72}
}

result = {}

for name, subjects in students.items():
    total = 0
    for marks in subjects.values():
        total += marks

    percentage = total / len(subjects)
    result[name] = {"Total": total, "Percentage": percentage}

print(result)
```

```python
#Reverse Dictionary with Grouping
#Reverse a dictionary where values repeat.
data = {
    "apple": "fruit",
    "carrot": "vegetable",
    "banana": "fruit",
    "spinach": "vegetable"
}

reverse = {}

for key, value in data.items():
    if value not in reverse:
        reverse[value] = []
    reverse[value].append(key)

print(reverse)
```

```python
#Merge Dictionaries with Value Addition
#Merge dictionaries and add values of common keys.
d1 = {"a": 10, "b": 20, "c": 30}
d2 = {"b": 5, "c": 15, "d": 25}

merged = {}

for key in d1.keys() | d2.keys():
    merged[key] = d1.get(key, 0) + d2.get(key, 0)

print(merged)
```

```python
#Frequency Count of Words (Advanced Logic)
sentence = "python is easy and python is powerful"

freq = {}

for word in sentence.split():
    freq[word] = freq.get(word, 0) + 1

print(freq)
```

```python
#Dictionary Sorting by Value (Without lambda)
scores = {"A": 85, "B": 92, "C": 78, "D": 90}

sorted_scores = {}

for k, v in sorted(scores.items(), key=lambda x: x[1]):
    sorted_scores[k] = v

print(sorted_scores)
```

```python
#Find Common Keys with Same Values
d1 = {"a": 10, "b": 20, "c": 30}
d2 = {"a": 10, "b": 25, "c": 30}

common = {}

for key in d1:
    if key in d2 and d1[key] == d2[key]:
        common[key] = d1[key]

print(common)
```

```python
#Character Frequency (Case-Insensitive)
text = "Programming"

char_freq = {}

for ch in text.lower():
    char_freq[ch] = char_freq.get(ch, 0) + 1

print(char_freq)
```

```python
#Convert List of Tuples to Nested Dictionary
data = [
    ("HR", "Amit", 50000),
    ("HR", "Riya", 55000),
    ("IT", "Karan", 70000)
]

dept = {}

for d, name, salary in data:
    if d not in dept:
        dept[d] = {}
    dept[d][name] = salary

print(dept)
```
